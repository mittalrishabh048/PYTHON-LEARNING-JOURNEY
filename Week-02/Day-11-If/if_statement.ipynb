DAY 11:CONDITIONAL STATEMENTS:1."IF"

```python
a==2,4,6,8
b==1,3,5,7
#1.
if(a==2,4,6,8):
  print("EVEN")
#2.
if(b==1,3,5,7):
  print("ODD")

#**IN "if" STATEMENT,IF ANY CONDITION IS FALSE IT WILL SKIP THAT(LIKE BELOW).
#3.
if(not(a==2,4,6,8)):
  print("ODD")
#4.
if(not(b==1,3,5,7)):
  print("EVEN")

#HERE 3 AND 4 ARE SKIPPED BECAUSE THEY ARE NOT TRUE AND 5 IS EXECUTED AFTER.
#5.
if(a==2,4,6,8):
  print("EVEN")
```
