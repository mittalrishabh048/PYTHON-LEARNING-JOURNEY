```python
#MINI-PROJECT==>BUDGET HELPER.
#PRACTICE:ALL OPERATORS AND NESTED IF/ELIF
income=float(input("ENTER YOUR MONTHLY INCOME:"))
savings_target=float(input("ENTER YOUR SAVINGS TARGET:"))

savings_rate=savings_target/income

print("\nBUDGET ANALYSIS:")
print("Income:",income)
print("Target:",savings_target)
print("Required:",savings_rate)

if(savings_rate<=0.3):
  print("Excellent.")
elif(savings_rate<=0.5):
  print("Good.")
else:
  print("Tight Budget.")
```
