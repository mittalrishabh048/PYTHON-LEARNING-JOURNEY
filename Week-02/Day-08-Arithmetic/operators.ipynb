DAY 8:ARITHMETIC OPERATORS(+,-,*,/,**,%)

```python
a=10
b=5
print(a+b)
print(a-b)
print(a*b)
print(a/b)    #WHENEVER WE DIVIDE,IT ALWAYS GIVES "Float" VALUE.
print(a**b)   #** a ki power b(EXPONENTIAL)
print(a%b)  #**  "%" used to find remainder.
```
