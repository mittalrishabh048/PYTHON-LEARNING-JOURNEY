DAY 10:ASSIGNMENT(=,+=,-=,*=,/=,**=,%=) AND LOGICAL(not,and,or) OPERATORS

```python
#ASSIGNMENT OPERATORS.
num=10
a=int(input("enter any number:"))
'''
num+=a
num-=a
num*=a
num/=a
num**=a
'''
num%=a
print(num)
```

```python
#LOGICAL OPERATORS.
#1.'not' OPERATOR:
print(not True)
print(not False)
a=int(input("enter num1:"))
b=int(input("enter num2:"))
print(a>b,",",not(a>b))

#2.'and' OPERATOR:
#**IT GIVES TRUE VALUE WHEN BOTH THE CONDITIONS AROUND ARE TRUE.
val1=True
val2=False
val3=True
print("AND OPERATOR:",val1 and val2,",",val1 and val3)

#3.'or' OPERATOR:
#**IT GIVES TRUE VALUE IF EITHER ONLY ONE VALUE IS TRUE OR BOTH ARE TRUE.
VAL1=True
VAL2=False
VAL3=False
print("OR OPERATOR:",VAL1 or VAL2,",",VAL2 or VAL3,",",VAL1 or VAL3)
```
