DAY 36:FILE I/O
PYTHON(AS A LANGUAGE) CAN BE USED TO PERFORM OPERATIONS ON A FILE(READ AND WRITE DATA).

TYPES OF FILES:
1.TEXT FILES.
2.BINARY FILES.

```python
'''
SYNTAX:
WE HAVE TO OPEN A FILE BEFORE READING OR WRITING.
<f=open("file_name","mode")
f.close()>
*we should always close the file.

**MODES ARE THE KIND OF PURPOSE FOR WHICH WE ARE OPENING THE FILE.
THESE CAN BE:"r"(read),"w"(write),"a"(append),etc.
'''
f=open("myfile.txt","r")
data=f.read()
#or simply:print(f.read())
print(data)
f.close()
```

```python
#File Reading:
##1:f.read()=>reads entire file.
f=open("myfile.txt","r")
data=f.read(5)
#or simply:print(f.read())
print(data)

##2:f.readlines()=>reads one line at a time.
g=open("myfile.txt","r")
line1=g.readlines(1)
print(line1)
line3=g.readlines(3)
print(line3)
g.close()
```

```python
#File Writing:
'''
1:f=open("file_name","w")   w=write mode
    ==>f.write("new line")
       *IT OVERWRITES ENTIRE FILE.
'''
f=open("myfile.txt","w")
f.write("1\nx\ny\nz")
f.close()

'''
2:f=open("file_name","a")  a=append mode
    ==>f.write("new line")
       *IT ADDS NEW LINE TO FILE.
'''
g=open("myfile.txt","a")
g.write("\nzzzz")
g.close()
```

```python
#  'with syntax':
with open("myfile.txt","r") as f:
  data=f.read()
  print(data)

#**While using 'with' we don't have to close file.

with open("myfile.txt","a") as g:
  g.write("\nNew Line")
```

```python
'''
DELETING A FILE:
Using the OS Module.
Module(like a code library) is a file written by another
programmer that generaly has a function that we can use.

Syntax:
import os
os.remove(file_name)
'''
import os
os.remove("demo.txt")
```
