DAY 25:SETS;
SET IS A COLLECTION OF UNORDERED ITEMS.
*EACH ELEMENT IN THE SET MUST BE IMMUTABLE/HASHABLE(int,float,str,tuples,etc.) AND UNIQUE.

```python
set={1,2,3,1,5,"hello","hello","world"}
print(type(set))
print(len(set))
print(set)
#*Ignores repeated elememts(as seen).
```

SET METHODS:

```python
#1."set.add(el)"(adds an element):
set={1,2}
set.add(3)
set.add((2,3,4))
set.add("str")
print(set)
```

```python
#2."set.remove(el)"(removes an element):
set={1,4,2,7}
set.remove(4)
set.remove(7)
print(set)
```

```python
#3."set.clear()"(empties the set):
set={2,3,45,"x","a",56.3,(2,5,6)}
print(len(set))
set.clear()
print(set)
print(len(set))
```

```python
#4."set.pop()"(removes a random value):
set={3,5,1,8,'abc'}
print(set.pop())
print(set.pop())
print(set)
```

```python
#5."set1.union(set2)"(combines two set values and returns new):
set1={2,4,5,"hello"}
set2={4,"hello","world"}
print(set1.union(set2))
print(set2.union(set2))
```

```python
#6."set1.intersection(set2)"(combines common values and returns new):
set1={2,4,5,"hello"}
set2={4,'hello'}
print(set1.intersection(set2))
print(set2.intersection(set1))
```
