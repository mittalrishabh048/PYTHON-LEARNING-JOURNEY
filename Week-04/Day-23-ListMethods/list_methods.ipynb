DAY 23:LIST METHODS

```python
#1."list.append()"(it adds one element at the end of list):
list=[2,1,3]
list.append(4)
print(list)
list.append(10)
print(list)
```

```python
#2."list.sort()"(it sorts in ascending order):
list=[2,1,3,10,7]
print(list.sort())#==>None
print(list)
```

```python
#3."list.sort(reverse=True)"(it sorts list in descending order):
list=[2,1,3,10,7]
print(list.sort(reverse=True))#==>None
print(list)
```

```python
##*SORTING ALSO WORKS ON STRINGS:
list=["a","d","f","b","k"]
list.sort()
print(list)
list.sort(reverse=True)
print(list)

fruits=["banana","kiwi","apple","orange"]
fruits.sort()
print(fruits)
fruits.sort(reverse=True)
print(fruits)
```

```python
#4."list.reverse()"(reverses list):
list=[7,4,89,34,98]
list.reverse()
print(list)
```

```python
#5."list.insert(idx,el)"(inserts element at the mentioned index)
list=[2,1,3,8,4]
list.insert(3,89)
print(list)
list.insert(0,5)
print(list)
```

```python
#6."list.remove()"(it removes first occurence of the element)
list=[2,1,3,8,4,2]
list.remove(1)
print(list)
list.remove(2)
print(list)
```

```python
#7."list.pop(idx)"(it removes the element from the mentioned index)
list=[2,1,3,8,4,2]
list.pop(3)
print(list)
list.pop(4)
print(list)
```
