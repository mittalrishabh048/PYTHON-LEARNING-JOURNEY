DAY 26:DICTIONERIES;
DICTIONERIES ARE USED TO STORE DATA VALUES IN "key:value" PAIRS.
*THEY ARE UNORDERED,MUTABLE(CHANGEABLE) AND DON'T ALLOW DUPLICATE KEYS.

```python
info={
    "keys":"values",
    "name":"x",
    "learning":"coding",
    "age":18,
    "is_adult":True,
    11:99,  #*KEYS CAN BE STR,INT OR FLOAT.
    1.0:99.0,
    "list":[1,3,5], #*WE CAN ADD LIST IN DICTIONERIES.
    "tup":(1,3,5) #*WE CAN ADD TUPLES IN DICTIONERIES.
}
info["surname"]="y"
print(info)
print(type(info))
print(info["name"])
print(info[11])
print(info[1.0])
print(info["tup"])
print(info["list"])
info["name"]="z"
print(info["name"])
```

```python
#WE CAN ALSO CREATE AN EMPTY DICT:
null_dict={}
print(null_dict)
null_dict["name"]="xyz"
print(null_dict)
```

```python
#NESTING IN DICTIONERIES:
dict={
    "name":"rahul",
    "subjects":{
        "physics":99,
        "chemistry":97,
        "maths":95
    }
}
print(dict)
print(dict["subjects"])
print(dict["subjects"]["maths"])
print(dict["subjects"]["physics"])
```
