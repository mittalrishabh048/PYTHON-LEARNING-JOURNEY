DAY 22:LIST(A BUILT-IN DATA TYPE THAT STORES SET OF VALUES)
*IT CAN STORE ELEMENTS OF DIFFERENT TYPES(INTEGER,FLOAT,STRING)

```python
marks=[78,90,99.9,67,55]
print(marks)
print(len(marks))
print(type(marks))
print(marks[0],marks[4])
```

```python
#WE CAN STORE DIFFERENT DATA TYPES IN A 'SINGLE' LIST.
student=["name",18,99.9]
print(student)
print(len(student))
print(type(student))
print(type(student[0]))
print(len(student[0]))
print(type(student[1]))
print(type(student[2]))
```

```python
#LIST SLICING(SIMILAR TO LIST SLICING)
marks=[78,90,99.8,67,55]
print(marks[ :4])
print(marks[1: ])
print(marks[1:len(marks)])

#*negative slicing:
print(marks[-3:-1])
```
