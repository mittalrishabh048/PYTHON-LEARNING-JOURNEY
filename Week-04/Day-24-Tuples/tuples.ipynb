DAY 24:TUPLES(A BUILT-IN DATA TYPE THAT LETS US CREATE IMMUTABLE SEQUENCE OF VALUE)

```python
tup=(2,1,4,7,3)
print(tup[0])
print(tup[3])
print(type(tup))

#NOT ALLOWED IN TUPLES:"tup[0]=2"
```

```python
tup=() #EMPTY TUPLE
print(tup)
print(type(tup))
```

```python
#"," is necessary for a single element tuple.
tup=(1,)
print(type(tup))
print(tup)

#otherwise(without ",");
tup1=(1)
print(tup1)
print(type(tup1))
```

**SLICING OF TUPLES IS SAME AS LIST SLICING

TUPLE METHODS:

```python
#1."tup.index(el)"(returns index of first occurence of elements):
tup=(2,1,3,1)
##BOTH THE WAYS ARE CORRECT:
tup.index(1)
   #OR;
print(tup.index(2))
```

```python
#2."tup.count(el)"(counts total occurence of elements):
tup=(2,1,3,1)
##BOTH THE WAYS ARE CORRECT:
tup.count(2)
   #OR;
print(tup.count(1))
```
