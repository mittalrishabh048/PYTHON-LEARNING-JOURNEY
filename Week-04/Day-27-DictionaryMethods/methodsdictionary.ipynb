  DAY 27:DICTIONARY METHODS

```python
student={
    "name":"rahul",
    "subjects":{
        "physics":90,
        "chemistry":99,
        "maths":89
    }
}

#1."dict.keys()"(returns all keys):
print(student.keys())
print(list(student.keys())) #type casting in dictionaries.
print(len(list(student.keys())))

#2."dict.values()"(returns all values):
print(student.values())
print(tuple(student.values()))
print(len(tuple(student.values())))

#3."dict.items()"(returns all "key:value" pairs):
print(student.items())
print(list(student.items()))
pairs=list(student.items())
print(pairs[0])
print(pairs[1])

#4."dict.get("key")"(returns the value according to key):
print(student.get("name"))
print(student.get("subjects"))
print(student.get("surname"))#it will give none.

#5."dict.update(newDict)"(insert the specified items in dictionary):
student.update({"surname":"kumar"})#1.
print(student)
new_dict={"x":"y"}#2.
student.update(new_dict) #Both(1 & 2) ways are correct.
print(student)
```
