DAY 19:NESTED LOOPS

```python
#SQUARE PATTERN:
for i in range(4):
  for j in range(4):
    print("*",end="")
  print()
```

```python
#NUMBER GRID:
for i in range(3):
  for j in range(3):
    print(j+1,end="")
  print()
```

```python
#MULTIPLICATION TABLE:
for i in range(1,4):
  for j in range(1,6):
    print(i*j,end="")
  print()
```

```python
#COORDINATE PAIRS:
for i in range(1,4):
  for j in range(1,4):
    print("(",i,",",j,")",end="")
  print()
```

```python
#ALPHABET GRID:
for i in range(4):
  for j in range(4):
    print(chr(65+j),end="")
  print()
```
