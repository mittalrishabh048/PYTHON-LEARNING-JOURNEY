DAY 20:PATTERN PRACTICE

```python
#REVERSE TRIANGLE:
for i in range(5,0,-1):
  print("*"*i)
```

```python
#RIGHT TRIANGLE:
for i in range(1,6):
  print("*"*i)
```

```python
#NUMBER TRIANGLE:
for i in range(1,6):
  print(str(i)*i)
```

```python
#HALF-PYRAMID TRIANGLE:
for i in range(1,6):
  for j in range(1,i+1):
    print(j,end="")
  print()
```

```python
#STAR PYRAMID:
rows=5
for i in range(1,rows+1):
  print(""*(rows-1)+"*"*(2*i))
```
