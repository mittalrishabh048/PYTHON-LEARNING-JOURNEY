DAY 16:RANGE FUNCTION
       *IT RETURNS A SEQUENCE OF NUMBERS.

```python
#1."range(stop):"
for i in range(5):
  print(i)
for i in range(10):
  print(i)

#2."range(start,stop):"
for i in range(2,8):
  print(i)
for i in range(4,11):
  print(i)

#3."range(start,stop,step):"
for i in range(2,10,3):
  print(i)
for i in range(10,101,25):
  print(i)

#PRACTICE QUES:PRINT EVEN AND ODD NUMBERS.
print("EVEN NUMBERS:")
even=range(2,100,2)
for i in even:
  print(i)

print("ODD NUMBERS:")
odd=range(1,100,2)
for i in odd:
  print(i)
```
