DAY 17:"while" LOOP.

```python
count=1
while count<=5:
  print(count)
  count+=1
print("final count:",count)
```

```python
i=1
while i<=100:  #i can be even 1000000(infinite)
  print("I=",i)
  i+=1
```

```python
#PRINT NUMBERS FROM 1 TO 5:
i=1
while i<6:
  print(i)
  i+=1
```

```python
#PRINT NUMBERS FROM 5 TO 1:
i=5
while i>0:
  print(i)
  i-=1
```

```python
#PRINT THE MULTIPLICATION OF ANY NUMBERS "N":
n=int(input("ENTER THE NUMBER:"))
i=1
while i<11:  #or i<=10
  print(n*i)
  i+=1
```
