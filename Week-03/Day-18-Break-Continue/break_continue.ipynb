DAY 18:*BREAK AND *CONTINUE.
       BREAK:USED TO DETERMINE THE LOOP WHEN ENCOUNTERED.
       CONTINUE:IT TERMINATES EXECUTION IN THE CURRENT ITERATION AND CONTINUES EXECUTION OF THE LOOP WITH THE NEXT ITERATION.


```python
i=1
while i<=5:
  print(i)
  if(i==4):
    break
  i+=1
print("END OF LOOP.")
```

```python
nums=(1,4,9,16,25,36,49,64,81,100,49)
x=49
i=0  #idx
while i<len(nums):
  if(nums[i]==x):
    print("FOUND AT idx",i)
    break
  else:
    print("finding..")
  i+=1

print("END OF LOOP.")
```

```python
i=0
while i<=5:
  if(i==3):
    i+=1
    continue
  print(i)
  i+=1
```
