DAY 33:RECURSIONS(WHEN FUNCTION CALLS ITSELF REPEATEDLY)

*LOOPS AND RECURSIONS ARE VERY SIMILAR.IN SOME CASES,WE PREFER USING LOOPS AND IN SOME CASES,WE PREFER USING RECURSIONS.

```python
#recursive functions:
#1.
def show(n1):
  if (n1==0): #*  =>base case(stopping value).
    return    # without base case value program will run infinite times like loops.
  print(n1)
  show(n1-1)

show(5)

#2.
def fact(n2):
  if (n2==0 or n2==1):
    return 1
  else:
    return n2*fact(n2-1)

print(fact(5))
print(fact(0))
print(fact(1))
print(fact(10))
```
