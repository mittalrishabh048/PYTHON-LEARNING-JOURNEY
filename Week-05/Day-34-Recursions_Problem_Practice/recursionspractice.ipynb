```python
#1.WRITE A RECURSIVE FUNCTION TO CALCULATE THE SUM OF FIRST n NUMBERS:
def numbers(n):
  if(n==0):
    return 0
  return numbers(n-1)+1

sum=numbers(10)
print(sum)
```

```python
'''
2.WRITE A RECURSIVE FUNCTION TO PRINT ALL THE ELEMENTS OF A LIST.
(HINT:USE LIST AND INDEX AS PARAMETERS):
'''
el_list=[1,"a",99.9,"b"]

def print_list(list,idx=0):
  if (idx==len(list)):
    return
  print(list[idx])
  print_list(list,idx+1)

print_list(el_list)
```
