DAY 30:PARAMETERS AND ARGUMENTS.

```python
'''
SYNTAX:
*funtion definition.
def funtion_name(a,b):  *here,a and b are parameters.
  code

calling function:
funtion_name(1,2)  *here,1 and 2 are arguments.
'''
```

```python
#1.SUM OF TWO NUMBER:
def calculate_sum(a,b):
  sum=a+b
  print(sum)

calculate_sum(2,5)
calculate_sum(67,45)

#2.AVG OF THREE NUMBERS:
def avg(a,b,c):
  average=(a+b+c)/3
  print(average)

avg(1,2,3)
avg(12,24,36)
```

```python
#DEFAULT PARAMETERS:
def calculate_product(a=2,b=2):
  product=a*b
  print(product)

calculate_product()
calculate_product(4,6)
```
