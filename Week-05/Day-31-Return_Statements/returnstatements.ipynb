```python
def sum(a,b):
  cal_sum=a+b
  return cal_sum

s=sum(3,7)
print(s)
```

```python
def sum(a,b):
  return 5  # only first return statement is executed.
  return a+b

s=sum(3,7)
print(s)
```

```python
#RETURN CUBE OF NUMBERS:
n=int(input("Enter number:"))
def findcube(n):
  cube=(n**3)
  return cube

findcube(n)
c=findcube(n)
print(c)
```
