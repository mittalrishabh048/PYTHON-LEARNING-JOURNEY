DAY 29:FUNCTIONS(FUNCTIONS ARE BLOCK OF STATEMENTS THAT PERFORMS A SPECIFIC TASK)

```python
'''
SYNTAX:
def funtion_name():
  code

calling:
funtion_name()
'''
```

```python
def greet():
  print("Welcome")
#STILL IT WILL NOT EXECUTE BECAUSE WE HAVE TO MAKE "call" TO FUNVTION:
greet()#calling the funtion
```

```python
print("Start")

def greet():
  print("Welcome",input("Enter Your Name:"))
greet()

print("End")
```

```python
#SOME OTHER FUNTIONS:
def my_name():
  print("My name is __.")
my_name()


def school():
  print("I have completed my class 12")
school()

def line():
  print("----")
line()
print("Python")
line()
```
