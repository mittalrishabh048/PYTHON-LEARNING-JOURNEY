```python
#PRACTICE QUES:
#1.WAP TO PRINT THE LENGTH OF A LIST:
list1=["Chandigarh","New Delhi","Bangalore","Kolkata"]
list2=[45,3,6,23,456,89.0]
def print_len(list):
  print(len(list))
  return len(list)

print_len(list1)
a=print_len(list2)
print(a)
```

```python
#2.WAP TO PRINT THE ELEMENTS OF A LIST IN A SINGLE LINE:
list1=[1,2,"Hello","World",(2,5)]
print(list1)

def elements(list):
  print(list[0],end=",")
  print(list[1],end=",")
  print(list[2],end=",")
  print(list[3],end=",")
  print(list[4])

elements(list1)

#OR

def print_el(list):
  for i in list:
    print(i,end=",")

print_el(list1)
```

```python
#3.WAP TO FIND THE FACTORIAL OF NUMBER n:
def print_fact(n):
  fact=1
  for i in range(1,n+1):
    fact*=i
  print(fact)

print_fact(5)
print_fact(6)
```

```python
#4.WAP TO CONVERT USD TO INR(CONSIDERING 1_USD=85_INR):
usd_1=85
def convert(n):
  print(usd_1*n)

convert(20)
convert(50)

#OR

def converter(usd_val):
  inr_val=usd_val*85
  print(usd_val,"USD=",inr_val,"INR")

converter(20)
converter(50)
```

```python
#5.WAP TO INPUT NUMBERS AND FIND WHETHER THEY ARE EVEN OR ODD:
num=int(input("Enter The Number:"))
def find(num):
  remainder=num%2
  if (remainder==0):
    print("EVEN")
  else:
    print("ODD")

find(num)
```
