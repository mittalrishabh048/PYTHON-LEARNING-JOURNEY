```python
#TYPE CONVERSION.
'''
   NOTE THAT=>IF WE ADD AN INTEGER VALUE AND A FLOAT VALUE,
              THEN THE OUTPUT IS OF THAT TYPE WHICH IS SUPERIOR(GREATER).
             *we can't add a "float" or "int" to a string.
              we can only add a str in another str.
'''

a=4
b=2.26
sum=a+b
print(sum)
print(type(sum))

c=2
d=5
sub=d-c
print(sub,",",type(sub))
```

```python
#TYPE CASTING.
a=int("5")
b=float("3.5")
c=int(2.5)
print(type(a),",",type(b),",",type(c))
sum1=a+b
sum2=a+c
print(sum1,",",sum2)
print(type(sum1),",",type(sum2))
```
