DAY 4:DATA TYPES

*THERE ARE 5 TYPES OF DATA TYPES:
1.INTEGERS(-VE,0,+VE)
2.STRING(MAINLY ALPHABETS THAT AE WRITTEN IN SINGLE OR DOUBLE OR TRIPLE QUOTES).
3.FLOAT(DECIMAL VALUE)
4.BOOLEAN(True OR False)
5.NONE

```python
name="rishabh"
age=17
price=99.99

print(type(name),type(age),type(price))

name1='rm'
name2="rm"
name3='''rm'''
print(type(name3),type(name1))

value1=True
value2=False
#**T in "True" and F in "False" should always be capital,otherwise error.
print(type(value1),type(value2))

a=None
#**N in "None" should be always capital otherwise error.
print(type(a))

'''
*THERE ARE SOME KEYWORDS(KIND OF SPECIAL WORDS "IN PYTHON")
 LIKE True,False,None MENTIONED ABOVE.
*WE WILL KNOW OTHER WORDS AS WE MOVE FORWARD.
'''
```
